## Importing Data

```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns

mp.__version__
```

## Set Style

```python
sns.set_style("whitegrid")
```

## Analyzing Data

```python
data = pd.read_csv('cost-of-living_v2.csv')
countries = ['China', 'India', 'United States', 'Brazil', 'Russia', 
             'United Kingdom', 'Germany', 'France', 'Japan', 'South Africa', 
             'Canada', 'Australia', 'Mexico', 'South Korea', 'Indonesia', 
             'Italy', 'Spain', 'Turkey', 'Nigeria', 'Saudi Arabia']

# gets rows of the countries within the countries array
d5 = data[data['country'].isin(countries)]

# Testing for 10 rows of data (speed)
#d5 = data.drop(data.index[10:4956])

# Maps each country to a continent
continent_map = {
    'China': 'Asia',
    'India': 'Asia',
    'United States': 'North America',
    'Brazil': 'South America',
    'Russia': 'Asia', 
    'United Kingdom': 'Europe', 
    'Germany': 'Europe', 
    'France': 'Europe', 
    'Japan': 'Asia', 
    'South Africa': 'Africa', 
    'Canada':'North America',
    'Australia': 'Australia', 
    'Mexico': 'North America',
    'South Korea': 'Asia', 
    'Indonesia': 'Asia', 
    'Italy': 'Europe', 
    'Spain': 'Europe', 
    'Turkey': 'Europe', 
    'Nigeria': 'Africa', 
    'Saudi Arabia': "Asia"
}
d5['continent'] = d5['country'].map(continent_map)

# Sorting Data example (by Apples)
# sorted_d5 = d5.sort_values(by=['x16'], na_position='first')
# sorted_d5

#prints new dataset
d5
```

## Income

```python
# Montly income salary plot
plt.figure(figsize=(24,10))
sns.scatterplot(data = d5, x = 'country', y= 'x54', hue = 'country')
plt.title('Monthly Salary Income per Country', fontsize = 25, fontweight = 800)
plt.ylabel("Monthly Salary Income", fontsize = 15)
plt.xlabel("Cities", fontsize = 15)

plt.legend(title = 'Country', bbox_to_anchor=(1.1, 1))
```

```python
# Plots the density of Monthly Salary Income
plt.figure(figsize=(10,8))
plt.title('Density for Monthly Salary Income per Country', fontsize = 20, fontweight = 800)
plt.xlabel('Monthly Salary (USD)', fontsize = 15)
plt.ylabel('Density', fontsize = 15)
sns.kdeplot(x=d5['x54'], hue = d5['country'], fill = True, alpha = 0.8, common_norm=False)
```

```python
# City count
plt.figure(figsize=(24,10))
g = sns.countplot(data = d5,x = 'country')
plt.title('Number of Cities in each Country', fontsize = 25, fontweight = 800)
plt.xlabel('Countries', fontsize = 15)
plt.ylabel('Number of Cities', fontsize = 15)

g.bar_label(g.containers[0])
```
